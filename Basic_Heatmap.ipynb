## Loading the dataset from a link

```r
url = "https://raw.githubusercontent.com/ahmedmoustafa/Notebooks/master/data/usa_monthly_temperature.tsv"
data = read.table(url, row.names = 1, header = TRUE)
```

```r
dim(data)
```

```r
head(data)
```

```r
tail(data)
```

## Converting the `data.frame` into a `matrix`

```r
data2 = as.matrix(data)
head(data2)
```

```r
heatmap(data2)
```

```r
heatmap(data2, scale = "none")
```

## Installing a Package (`gplots`)

```r
install.packages ("gplots")
```

```r
library(gplots)
```

```r
heatmap(data2, scale = "none", col = redblue(256))
```

```r
heatmap(data2, scale = "none", col = rev(redblue(256)))
```
